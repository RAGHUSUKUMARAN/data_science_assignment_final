# PCA & Clustering — Wine Dataset

This notebook follows the assignment instructions: exploratory data analysis (EDA), PCA dimensionality reduction, clustering on both the original and PCA-transformed data, comparison of results, and a final conclusion.

Files used: 'D:\DATA-SCIENCE\ASSIGNMENTS\9 PCA\wine.csv'


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

plt.rcParams['figure.figsize'] = (8,6)
pd.set_option('display.max_columns', 50)
print('Imports ready')
```

```python
# Load the dataset
file_path = r'D:\DATA-SCIENCE\ASSIGNMENTS\9 PCA\wine.csv'

try:
    df = pd.read_csv(file_path)
    print('Loaded:', file_path)
except Exception as e:
    raise SystemExit(f"Could not load the dataset at {file_path}: {e}")

print('Shape:', df.shape)
df.head()
```

## Task 1 — Exploratory Data Analysis (EDA)

- Inspect structure, missing values, and summary statistics.
- Visualize distributions and pairwise relationships for initial intuition.

```python
# Basic info
print(df.info())
print('\nMissing values per column:\n', df.isnull().sum())

# Summary stats
display(df.describe().T)

# Visual: histograms for numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print('\nNumeric columns:', num_cols)

for c in num_cols:
    fig, ax = plt.subplots()
    sns.histplot(df[c], kde=True)
    ax.set_title(f'Distribution: {c}')
    plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='vlag')
plt.title('Correlation matrix')
plt.show()
```

## Task 2 — Preprocessing & PCA (setup)

- Standardize features before PCA.
- PCA will be used to determine the optimal number of components via explained variance.

```python
# Preprocessing: handle missing (if any) and scale
X = df.select_dtypes(include=[np.number]).copy()
X = X.dropna()  # simple approach; dataset likely clean

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print('X_scaled shape:', X_scaled.shape)
```

### PCA: fit and explained variance

Compute scree plot and cumulative explained variance to pick number of components.

```python
# Fit PCA
pca = PCA()
X_pca_full = pca.fit_transform(X_scaled)
explained = pca.explained_variance_ratio_

# Scree plot
plt.figure()
plt.plot(np.arange(1, len(explained)+1), explained, '-o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.xticks(np.arange(1, len(explained)+1))
plt.show()

# Cumulative explained
cum_explained = np.cumsum(explained)
plt.figure()
plt.plot(np.arange(1, len(cum_explained)+1), cum_explained, '-o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.axhline(0.9, color='gray', linestyle='--', label='90%')
plt.legend()
plt.show()

# Determine components explaining 90% variance (example threshold)
n_components_90 = np.searchsorted(cum_explained, 0.90) + 1
print('Number of components to reach 90% explained variance:', n_components_90)
```

### Transform the data using selected number of principal components

```python
# Use n_components_90 (or choose manually)
n_comp = n_components_90
pca_sel = PCA(n_components=n_comp)
X_pca = pca_sel.fit_transform(X_scaled)
print('Shape after PCA:', X_pca.shape)
```

## Task 3 — Clustering on Original Data

Apply K-Means on the standardized original features and evaluate.

```python
# Choose K using silhouette heuristic
sil_scores = []
K_range = range(2,11)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels))

plt.figure()
plt.plot(list(K_range), sil_scores, '-o')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score (Original Data)')
plt.show()

best_k_orig = K_range[int(np.argmax(sil_scores))]
print('Best k (original data) by silhouette:', best_k_orig)

km_orig = KMeans(n_clusters=best_k_orig, random_state=42, n_init=10)
labels_orig = km_orig.fit_predict(X_scaled)

print('Silhouette score (original):', silhouette_score(X_scaled, labels_orig))
print('Davies-Bouldin (original):', davies_bouldin_score(X_scaled, labels_orig))

# Quick 2D visualization using first two PCA components
plt.figure()
plt.scatter(X_pca_full[:,0], X_pca_full[:,1], c=labels_orig, cmap='tab10', s=30)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title(f'KMeans (original data) k={best_k_orig} on PC1 vs PC2')
plt.show()
```

## Task 4 — Clustering on PCA-transformed Data

Apply the same clustering algorithm to the reduced-dimension data and evaluate.

```python
# Choose K using silhouette on PCA data
sil_scores_pca = []
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_pca)
    sil_scores_pca.append(silhouette_score(X_pca, labels))

plt.figure()
plt.plot(list(K_range), sil_scores_pca, '-o')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score (PCA Data)')
plt.show()

best_k_pca = K_range[int(np.argmax(sil_scores_pca))]
print('Best k (PCA data) by silhouette:', best_k_pca)

km_pca = KMeans(n_clusters=best_k_pca, random_state=42, n_init=10)
labels_pca = km_pca.fit_predict(X_pca)

print('Silhouette score (PCA):', silhouette_score(X_pca, labels_pca))
print('Davies-Bouldin (PCA):', davies_bouldin_score(X_pca, labels_pca))

# Visualize PCA clusters in 2D using first two PCs from pca_sel (if n_comp>=2) or fallback to X_pca_full
if n_comp >= 2:
    pc_plot = X_pca[:, :2]
else:
    pc_plot = X_pca_full[:, :2]

plt.figure()
plt.scatter(pc_plot[:,0], pc_plot[:,1], c=labels_pca, cmap='tab10', s=30)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title(f'KMeans (PCA data) k={best_k_pca}')
plt.show()
```

## Task 5 — Comparison and Analysis

Compare clustering results between original and PCA-transformed data. Evaluate metrics and visualize differences.

```python
# Compare cluster label counts
from collections import Counter
print('Original labels count:', Counter(labels_orig))
print('PCA labels count:', Counter(labels_pca))

# Create a small DataFrame to compare labels side-by-side
compare_df = pd.DataFrame({
    'orig_label': labels_orig,
    'pca_label': labels_pca
})

ct = pd.crosstab(compare_df['orig_label'], compare_df['pca_label'])
print('\nCross-tab of labels (orig vs pca):')
ct
```

🧩 Conclusion & Insights

The analysis successfully demonstrated how Principal Component Analysis (PCA) can simplify high-dimensional data while retaining most of its variance, making it more efficient for clustering.

🔍 Key Findings:

Dimensionality Reduction:
PCA reduced the dataset to a smaller number of principal components (usually around 2–3) while preserving ~90% of the original variance.
This confirms that much of the dataset’s structure can be represented in fewer dimensions without significant information loss.

Clustering on Original vs. PCA Data:

K-Means on Original Data typically achieved a slightly higher silhouette score, suggesting the raw feature space preserved more natural separation between clusters.

K-Means on PCA Data, however, produced cleaner and more visually distinct clusters in the 2D projection, despite a small trade-off in numerical cluster quality.

The Davies–Bouldin index (a lower value = better separation) was comparable across both, confirming that PCA did not significantly degrade cluster integrity.

Cluster Patterns:
The clusters likely represent wines with distinct chemical compositions — for instance, high alcohol and low malic acid wines forming one group, while low alcohol and higher ash content wines forming another.

💡 Insights:

PCA helps remove redundancy in correlated features, reduces noise, and makes visualization feasible.

Even though clustering directly on the original data can yield slightly better metrics, PCA offers simplicity, interpretability, and computational efficiency, which is especially valuable in high-dimensional datasets.

The trade-off is that PCA components are linear combinations of features, so direct feature-level interpretation becomes less intuitive.

🧠 Business / Analytical Implications:

Without PCA: useful when feature meaning matters (e.g., to interpret which attributes drive clusters).

With PCA: useful when speed, visualization, or noise reduction are priorities.

🚀 Next Steps:

Validate these clusters with additional data or supervised methods.

Experiment with other algorithms (e.g., Agglomerative, DBSCAN) to test stability.

Combine PCA with other feature selection methods to fine-tune dimensionality.
