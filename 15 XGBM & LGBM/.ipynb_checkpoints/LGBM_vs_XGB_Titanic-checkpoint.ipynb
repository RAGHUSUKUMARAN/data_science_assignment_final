## 1) Data Loading
Load train and test CSV files. If local files are missing, the notebook will look for them in the sandbox `/mnt/data` as a fallback.

```python
# 1) Data Loading
import os, pandas as pd, numpy as np
DATA_DIR = r"D:\DATA-SCIENCE\ASSIGNMENTS\15 XGBM & LGBM"
train_path = os.path.join(DATA_DIR, "Titanic_train.csv")
test_path = os.path.join(DATA_DIR, "Titanic_test.csv")

# Fallback to sandbox paths if local not found
if not os.path.exists(train_path):
    train_path = "/mnt/data/Titanic_train.csv"
if not os.path.exists(test_path):
    test_path = "/mnt/data/Titanic_test.csv"

print("Train path:", train_path)
print("Test path:", test_path)

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

print("Train shape:", df_train.shape)
print("Test shape:", df_test.shape)
display(df_train.head())
```

## 2) Exploratory Data Analysis (EDA)
Inspect structure, missing values, distributions, and relationships with target (`Survived`).

```python
# 2) EDA
import matplotlib.pyplot as plt, seaborn as sns
sns.set(style="whitegrid")

print("Columns:", df_train.columns.tolist())
print("\nMissing values (train):")
display(df_train.isnull().sum().to_frame('missing'))

# Basic numeric summaries
display(df_train.describe(include='all').T)

# Survived distribution
plt.figure(figsize=(4,3))
sns.countplot(x='Survived', data=df_train)
plt.title('Survived distribution (train)')
plt.show()

# Age distribution by survival
if 'Age' in df_train.columns:
    plt.figure(figsize=(6,3))
    sns.kdeplot(df_train.loc[df_train['Survived']==0,'Age'].dropna(), label='Not Survived')
    sns.kdeplot(df_train.loc[df_train['Survived']==1,'Age'].dropna(), label='Survived')
    plt.title('Age distribution by survival')
    plt.legend()
    plt.show()

# Pclass vs Survived
if 'Pclass' in df_train.columns:
    plt.figure(figsize=(5,3))
    sns.barplot(x='Pclass', y='Survived', data=df_train)
    plt.title('Survival rate by Pclass')
    plt.show()

# Sex vs Survived
if 'Sex' in df_train.columns:
    plt.figure(figsize=(5,3))
    sns.barplot(x='Sex', y='Survived', data=df_train)
    plt.title('Survival rate by Sex')
    plt.show()
```

## 3) Preprocessing
Handle missing values, create useful features, encode categoricals, and prepare datasets for modeling. We'll keep the pipeline simple and reproducible.

```python
# 3) Preprocessing & Feature Engineering
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Work on copies
df = df_train.copy()

# Basic feature engineering: Title from Name, FamilySize from SibSp+Parch, IsAlone
def extract_title(name):
    try:
        return name.split(',')[1].split('.')[0].strip()
    except Exception:
        return "None"

if 'Name' in df.columns:
    df['Title'] = df['Name'].apply(extract_title)
    # group rare titles
    title_counts = df['Title'].value_counts()
    rare_titles = title_counts[title_counts < 10].index.tolist()
    df['Title'] = df['Title'].replace(rare_titles, 'Rare')

# Family size features
if set(['SibSp','Parch']).issubset(df.columns):
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

# Drop columns that won't be used directly
drop_cols = []
for c in ['Ticket','Cabin','Name']:
    if c in df.columns:
        drop_cols.append(c)

df = df.drop(columns=drop_cols, errors='ignore')

# Target and features separation
target_col = 'Survived'
y = df[target_col]
X = df.drop(columns=[target_col])

# Identify numeric and categorical columns
num_cols = X.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object','category']).columns.tolist()

print('Numeric cols:', num_cols)
print('Categorical cols:', cat_cols)

# Simple imputers and encoders
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

preprocessor = ColumnTransformer(transformers=[
    ('num', num_imputer, num_cols),
    ('cat', Pipeline(steps=[('impute', cat_imputer), ('ohe', ohe)]), cat_cols)
], remainder='drop')
```

### Train/validation split (from the original training file)
We'll split the original training set into training + validation for model comparison and tuning.

```python
# 3b) Train/validation split
from sklearn.model_selection import train_test_split
X_full = X.copy()
y_full = y.copy()
X_train, X_val, y_train, y_val = train_test_split(X_full, y_full, test_size=0.2, random_state=42, stratify=y_full)
print('Shapes:', X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

## 4) Model Pipelines â€” LightGBM & XGBoost
We will create pipelines that include preprocessing and the estimator. If LightGBM or XGBoost is missing, the cells will print instructions to install.

```python
# 4) Pipelines and model definitions
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Try to import LightGBM and XGBoost
lgb_installed = True
xgb_installed = True
try:
    from lightgbm import LGBMClassifier
except Exception as e:
    lgb_installed = False
    print("LightGBM not installed. Install with: pip install lightgbm")

try:
    from xgboost import XGBClassifier
except Exception as e:
    xgb_installed = False
    print("XGBoost not installed. Install with: pip install xgboost")

# Create pipeline function
def build_pipeline(estimator):
    return Pipeline(steps=[('pre', preprocessor), ('clf', estimator)])

# Default estimators
if lgb_installed:
    lgb_default = LGBMClassifier(random_state=42)
else:
    lgb_default = None

if xgb_installed:
    xgb_default = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
else:
    xgb_default = None

print("Pipelines ready.")
```

## 5) Train baseline models
Train default LightGBM and XGBoost pipelines and evaluate on validation set.

```python
# 5) Train baseline models (if installed)
results = {}
if lgb_default is not None:
    pipe_lgb = build_pipeline(lgb_default)
    pipe_lgb.fit(X_train, y_train)
    y_pred_lgb = pipe_lgb.predict(X_val)
    results['LGBM'] = {
        'accuracy': accuracy_score(y_val, y_pred_lgb),
        'precision': precision_score(y_val, y_pred_lgb, zero_division=0),
        'recall': recall_score(y_val, y_pred_lgb, zero_division=0),
        'f1': f1_score(y_val, y_pred_lgb, zero_division=0)
    }
    print('LGBM baseline metrics:', results['LGBM'])

if xgb_default is not None:
    pipe_xgb = build_pipeline(xgb_default)
    pipe_xgb.fit(X_train, y_train)
    y_pred_xgb = pipe_xgb.predict(X_val)
    results['XGB'] = {
        'accuracy': accuracy_score(y_val, y_pred_xgb),
        'precision': precision_score(y_val, y_pred_xgb, zero_division=0),
        'recall': recall_score(y_val, y_pred_xgb, zero_division=0),
        'f1': f1_score(y_val, y_pred_xgb, zero_division=0)
    }
    print('XGB baseline metrics:', results['XGB'])

display(results)
```

## 6) Hyperparameter Tuning (RandomizedSearchCV)
We'll run randomized search for a few key hyperparameters to keep runtime reasonable. Use GridSearch for exhaustive tuning if needed.

```python
# 6) Hyperparameter tuning with RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
import scipy.stats as st

tuned_results = {}

if lgb_default is not None:
    lgb_pipe = build_pipeline(LGBMClassifier(random_state=42))
    param_dist_lgb = {
        'clf__n_estimators': [100,200,400],
        'clf__num_leaves': [15,31,63],
        'clf__learning_rate': [0.01,0.05,0.1],
        'clf__max_depth': [3,5,7,9,None]
    }
    rsearch_lgb = RandomizedSearchCV(lgb_pipe, param_dist_lgb, n_iter=12, scoring='f1', cv=5, random_state=42, n_jobs=-1, verbose=1)
    rsearch_lgb.fit(X_train, y_train)
    tuned_results['LGBM'] = rsearch_lgb.best_params_
    best_lgb = rsearch_lgb.best_estimator_
    print('LGBM best params:', tuned_results['LGBM'])

if xgb_default is not None:
    xgb_pipe = build_pipeline(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42))
    param_dist_xgb = {
        'clf__n_estimators': [100,200,400],
        'clf__max_depth': [3,5,7,9],
        'clf__learning_rate': [0.01,0.05,0.1],
        'clf__subsample': [0.6,0.8,1.0]
    }
    rsearch_xgb = RandomizedSearchCV(xgb_pipe, param_dist_xgb, n_iter=12, scoring='f1', cv=5, random_state=42, n_jobs=-1, verbose=1)
    rsearch_xgb.fit(X_train, y_train)
    tuned_results['XGB'] = rsearch_xgb.best_params_
    best_xgb = rsearch_xgb.best_estimator_
    print('XGB best params:', tuned_results['XGB'])

display(tuned_results)
```

## 7) Evaluation on Validation and (optional) Test Set
Calculate metrics and plot comparisons. If `df_test` contains `Survived`, evaluate there too; otherwise, show validation performance and save predictions.

```python
# 7) Evaluate tuned models
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, f1_score
eval_table = {}

# Evaluate best_lgb and best_xgb if available
if 'best_lgb' in globals():
    y_pred = best_lgb.predict(X_val)
    eval_table['LGBM'] = {
        'accuracy': accuracy_score(y_val, y_pred),
        'f1': f1_score(y_val, y_pred, zero_division=0)
    }
    print("LGBM validation metrics:\n", classification_report(y_val, y_pred, zero_division=0))

if 'best_xgb' in globals():
    y_pred = best_xgb.predict(X_val)
    eval_table['XGB'] = {
        'accuracy': accuracy_score(y_val, y_pred),
        'f1': f1_score(y_val, y_pred, zero_division=0)
    }
    print("XGB validation metrics:\n", classification_report(y_val, y_pred, zero_division=0))

display(eval_table)

# If test file contains Survived, evaluate there too
if 'Survived' in df_test.columns:
    X_test_full = df_test.drop(columns=['Survived'], errors='ignore')
    # Apply same preprocessing via pipeline predict
    if 'best_lgb' in globals():
        y_test_pred_lgb = best_lgb.predict(X_test_full)
        print("LGBM test metrics (if ground truth available):")
        if 'Survived' in df_test.columns:
            print(classification_report(df_test['Survived'], y_test_pred_lgb, zero_division=0))
    if 'best_xgb' in globals():
        y_test_pred_xgb = best_xgb.predict(X_test_full)
        if 'Survived' in df_test.columns:
            print("XGB test metrics (if ground truth available):")
            print(classification_report(df_test['Survived'], y_test_pred_xgb, zero_division=0))
else:
    print("No 'Survived' column in test data â€” predictions will be saved instead.")
```

## 8) Comparison Plots
Visualize validation metrics side-by-side.

```python
# 8) Comparison plots
import pandas as pd
res_df = pd.DataFrame(eval_table).T
if not res_df.empty:
    res_df.plot(kind='bar', figsize=(6,4), ylim=(0,1))
    plt.title('Validation metrics comparison')
    plt.show()
else:
    print("No evaluation results to plot.")
```
