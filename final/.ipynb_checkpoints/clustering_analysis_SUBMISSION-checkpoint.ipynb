# ‚úàÔ∏è EastWest Airlines ‚Äî Customer Clustering Analysis

This project applies **unsupervised learning** techniques to the *EastWest Airlines* dataset in order to segment customers based on their travel and loyalty patterns.

Using **K-Means**, **Hierarchical (Agglomerative)**, and **DBSCAN** clustering, we identify meaningful customer groups that can help the airline optimize marketing, improve loyalty programs, and tailor services.

### üîç Objectives
- Perform **exploratory data analysis (EDA)** to understand the distribution and relationships among variables.  
- **Preprocess and scale** features to make them suitable for clustering.  
- Apply **multiple clustering algorithms** and compare their performance using silhouette scores.  
- Interpret and visualize cluster characteristics to understand customer behavior patterns.

---

üí° *Note:* Clustering is exploratory ‚Äî there‚Äôs no ‚Äúground truth.‚Äù The goal is to discover natural groupings within the data that make business sense.


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, silhouette_samples
from scipy.cluster import hierarchy

plt.rcParams['figure.figsize'] = (8,6)

# Display settings
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 20)
print('Imports ready')
```

```python
file_path = r'D:\DATA-SCIENCE\ASSIGNMENTS\8 clustering\EastWestAirlines.xlsx'
df = pd.read_excel(file_path)
print('Loaded:', file_path)
print('Shape:', df.shape)
df.head()
```

## Exploratory Data Analysis (EDA)

- Check for missing values, datatypes, and basic statistics.
- Visual exploration of distributions and pairwise relationships.

```python
# Basic info
print(df.info())

# Missing values
print('\nMissing values per column:\n', df.isnull().sum())

# Summary stats
df.describe().T
```

## Preprocessing

Steps:
1. Handle missing values (drop or impute depending on situation).
2. Remove or cap obvious outliers if necessary.
3. Scale features using `StandardScaler`.
4. Optionally use PCA for visualization.

```python
# --- Preprocessing ---
# For assignment generality we'll:
# - Drop rows with > 50% missing
# - Impute numeric NaNs with median
# - Encode categorical columns if present (one-hot)

# 1) Drop rows with many missing values
thresh = int(df.shape[1] * 0.5)
df_clean = df.dropna(thresh=thresh).copy()

# 2) Separate numeric and categorical
num_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df_clean.select_dtypes(include=['object','category']).columns.tolist()

# Impute numeric with median
for c in num_cols:
    if df_clean[c].isnull().any():
        df_clean[c].fillna(df_clean[c].median(), inplace=True)

# One-hot encode categoricals (if any)
if len(cat_cols) > 0:
    df_clean = pd.get_dummies(df_clean, columns=cat_cols, drop_first=True)

# Recompute numeric columns after encoding
num_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()

print('Clean shape:', df_clean.shape)
print('Numeric columns used for clustering:', num_cols)

# Keep a copy of the raw numeric dataframe used for clustering
X_raw = df_clean[num_cols].copy()

# 3) Scale
scaler = StandardScaler()
X = scaler.fit_transform(X_raw)

# Quick PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print('Explained variance by 2 PCA components:', pca.explained_variance_ratio_.sum())

# Attach PCA back to dataframe for plotting
plot_df = pd.DataFrame(X_pca, columns=['PC1','PC2'])
plot_df.index = df_clean.index
plot_df.head()
```

```python
# Plot distributions of first 6 numeric features (if available)
cols_to_plot = num_cols[:6]
for c in cols_to_plot:
    fig, ax = plt.subplots()
    sns.histplot(X_raw[c], kde=True)
    ax.set_title(f'Distribution: {c}')
    plt.show()

# PCA scatter
plt.figure(figsize=(8,6))
sns.scatterplot(x='PC1', y='PC2', data=plot_df)
plt.title('PCA projection (2 components)')
plt.show()
```

## K-Means Clustering

- Use the elbow method to choose `k`.
- Evaluate clusters using silhouette score.
- Visualize clusters on PCA components.

```python
# Elbow method + silhouette
inertia = []
sil_scores = []
K_range = range(2,11)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    inertia.append(km.inertia_)
    sil_scores.append(silhouette_score(X, labels))

plt.figure()
plt.plot(list(K_range), inertia, '-o')
plt.xlabel('k')
plt.ylabel('Inertia (within-cluster sum of squares)')
plt.title('Elbow Method for KMeans')
plt.show()

plt.figure()
plt.plot(list(K_range), sil_scores, '-o')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs k')
plt.show()

# Choose k with highest silhouette (example)
best_k = K_range[int(np.argmax(sil_scores))]
print('Suggested best k by silhouette:', best_k)

km = KMeans(n_clusters=best_k, random_state=42, n_init=10)
labels_km = km.fit_predict(X)

# Attach to plot_df
plot_df['kmeans_label'] = labels_km

sns.scatterplot(x='PC1', y='PC2', hue='kmeans_label', palette='tab10', data=plot_df)
plt.title(f'KMeans clusters (k={best_k})')
plt.legend(title='cluster')
plt.show()

print('Silhouette score for chosen k:', silhouette_score(X, labels_km))
```

## Hierarchical (Agglomerative) Clustering

- Plot a dendrogram and experiment with linkage methods.
- Cut the dendrogram to form clusters and compare with KMeans.

```python
# Hierarchical clustering - dendrogram
linkage_matrix = hierarchy.linkage(X[:200] if X.shape[0]>200 else X, method='ward')
plt.figure(figsize=(12,6))
hierarchy.dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title('Dendrogram (truncated)')
plt.xlabel('Sample index or (cluster size)')
plt.show()

# Choose number of clusters manually (e.g., same as best_k)
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=best_k, linkage='ward')
labels_ac = ac.fit_predict(X)
plot_df['agg_label'] = labels_ac

sns.scatterplot(x='PC1', y='PC2', hue='agg_label', palette='tab10', data=plot_df)
plt.title('Agglomerative Clustering labels')
plt.show()

print('Silhouette score (agglo):', silhouette_score(X, labels_ac))
```

## DBSCAN

- Density-based clustering that can find arbitrarily shaped clusters and mark outliers as `-1`.
- Important hyperparameters: `eps` and `min_samples`.

```python
# Simple heuristic to pick eps: k-distance plot
from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=5)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
distances = np.sort(distances[:,4])

plt.figure()
plt.plot(distances)
plt.ylabel('5-NN distance')
plt.xlabel('Points sorted by distance')
plt.title('k-distance plot for DBSCAN eps selection')
plt.show()

# Try couple of eps values
for eps in [0.5, 1.0, 1.5]:
    db = DBSCAN(eps=eps, min_samples=5)
    labels_db = db.fit_predict(X)
    n_clusters = len(set(labels_db)) - (1 if -1 in labels_db else 0)
    sil = silhouette_score(X, labels_db) if n_clusters > 1 else np.nan
    print(f'eps={eps} -> clusters={n_clusters}, silhouette={sil}')

# Choose an eps (example 0.5) and plot
db = DBSCAN(eps=0.5, min_samples=5)
labels_db = db.fit_predict(X)
plot_df['db_label'] = labels_db

sns.scatterplot(x='PC1', y='PC2', hue='db_label', palette='tab10', data=plot_df)
plt.title('DBSCAN labels')
plt.show()

print('DBSCAN unique labels (including -1 for noise):', np.unique(labels_db))
```

```python
# Profile clusters (KMeans)
profile = X_raw.copy()
profile['cluster_km'] = labels_km
profile.groupby('cluster_km').mean()

# Save a CSV of the profile for instructor review
save_path = r'D:\DATA-SCIENCE\ASSIGNMENTS\8 clustering\cluster_profile_km.csv'
profile.to_csv(save_path, index=False)
print(f'Saved cluster_profile_km.csv to {save_path}')
```
