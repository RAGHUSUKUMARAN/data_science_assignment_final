## Step 4: Data Cleaning & Outlier Detection

Before performing deeper visualizations, we must ensure the dataset is clean and free from inconsistencies.  
In this step, we:
1. Check for duplicate rows.  
2. Confirm that data types are correctly assigned.  
3. Visualize distributions to identify outliers.  
4. Use the **Interquartile Range (IQR)** rule to detect and optionally filter extreme values.  

Outliers can distort statistical measures and visual interpretations, so detecting them early ensures more reliable insights.


```python
# Step 4: Data Cleaning & Outlier Detection

# 1. Check for duplicated rows
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# 2. Verify data types (should all be numeric)
print("\n== Data Types ==")
print(df.dtypes)

# 3. Basic summary to confirm consistency
print("\n== Descriptive Summary ==")
display(df.describe().T)

# 4. Outlier visualization (boxplot for all numeric features)
plt.figure(figsize=(15,6))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title("Boxplot of All Numerical Features (Detecting Outliers)")
plt.show()

# 5. Optional: outlier detection using IQR rule
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outlier_mask = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR)))
outlier_count = outlier_mask.sum().sort_values(ascending=False)

print("\n== Outlier Counts per Column ==")
print(outlier_count[outlier_count > 0])

# 6. (Optional) Clean extreme outliers for visualization clarity (don’t delete data for real analysis)
df_clean = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f"\nRows before cleaning: {df.shape[0]} | After cleaning: {df_clean.shape[0]}")
```

**Observations:**
- No duplicate rows were found, indicating unique records.  
- All variables appear to be numeric, consistent with continuous signal readings.  
- Boxplots highlight a few extreme values in certain variables like decelerations (`DL`, `DP`) or variability metrics (`ASTV`, `ALTV`).  
- Using the IQR method, a small number of rows were identified as outliers and removed in the `df_clean` DataFrame for optional visual clarity.

## Step 6: Pattern Recognition and Insights

With a clean dataset and visualized distributions, we now identify key patterns and relationships.  
The main objectives here are:

1. To determine which features are most correlated with **Baseline Fetal Heart Rate (LB)**.  
2. To identify any strong positive or negative correlations that may have physiological significance.  
3. To visualize these relationships using scatter plots and correlation clustering.


```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform

# 1. Compute correlation matrix
corr_matrix = df_clean.corr()

# 2. Convert correlation to distance (1 - correlation)
dist_mat = 1 - corr_matrix

# 3. Replace any NaNs (in case of constant columns)
dist_mat = dist_mat.fillna(0)

# 4. Convert square matrix → condensed form for clustering
condensed = squareform(dist_mat.values)

# 5. Perform hierarchical clustering using Ward’s linkage
corr_linkage = hierarchy.linkage(condensed, method='ward')

# 6. Plot dendrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
hierarchy.dendrogram(corr_linkage, labels=corr_matrix.columns, leaf_rotation=90)
plt.title("Feature Correlation Clustering (Ward Linkage)")
plt.tight_layout()
plt.show()
```

**Observations & Insights:**
- Some features (like `MSTV` or `MLTV`) may show positive correlation with `LB`, indicating that higher variability is associated with a stable heart rate.
- Variables such as `DL`, `DS`, or `DP` (decelerations) may show mild negative correlations, suggesting that frequent decelerations coincide with lower baseline heart rates.
- The dendrogram groups related variables — for instance, `ASTV`, `ALTV`, and `MSTV` often cluster together, indicating they capture similar variability patterns.
- These relationships align with clinical expectations: healthy fetal monitoring patterns usually exhibit moderate variability and fewer prolonged decelerations.


```python
plt.figure(figsize=(12, 5))
dn = hierarchy.dendrogram(corr_linkage, labels=corr_matrix.columns,
                          leaf_rotation=90, color_threshold=None)

# color_threshold can be set to highlight clusters:
color_threshold = 0.7 * max(corr_linkage[:,2])  # example heuristic
plt.figure(figsize=(12, 5))
hierarchy.dendrogram(corr_linkage, labels=corr_matrix.columns, leaf_rotation=90,
                     color_threshold=color_threshold)
plt.title('Dendrogram with color threshold')
plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
import pandas as pd

# cut the dendrogram to form clusters
# try different numbers (3, 4, 5...) depending on how many distinct groups you see visually
k = 4  
cluster_labels = fcluster(corr_linkage, k, criterion='maxclust')

# create dataframe of feature -> cluster mappings
cluster_map = pd.DataFrame({
    'Feature': corr_matrix.columns,
    'Cluster': cluster_labels
}).sort_values('Cluster')

# print the cluster groups
for cluster in sorted(cluster_map['Cluster'].unique()):
    features = cluster_map[cluster_map['Cluster'] == cluster]['Feature'].tolist()
    print(f"\nCluster {cluster} ({len(features)} features):")
    print(", ".join(features))
```

### Step 8: Feature Reduction and Modeling Preparation

After identifying correlated feature clusters, the next goal is to **reduce redundancy** and **prepare the dataset for modeling**.  
Highly correlated features often convey overlapping information, which can distort model weights, inflate variance, and reduce interpretability.

#### Objectives:
1. **Minimize multicollinearity** — keep only the most informative or representative variables per cluster.  
2. **Simplify the feature space** — create a compact set of inputs for classification or predictive modeling.  
3. **Preserve interpretability** — retain features that have clear domain meaning or clinical relevance.

#### Methods:
- **Manual Feature Selection:**  
  From each cluster, pick one representative feature — ideally the one that is:
  - Most interpretable (clinically or statistically),
  - Most complete (fewest missing values), and
  - Least noisy (stable distribution).

- **Principal Component Analysis (PCA):**  
  Within each cluster, apply PCA to compress correlated variables into a single composite feature (the first principal component).  
  This captures most of the variance while removing redundancy.

#### Example Workflow:
1. Split the dataset into cluster-specific subsets using the cluster assignments from Step 7.  
2. Standardize each subset (mean = 0, variance = 1).  
3. Apply PCA within each subset to extract one component per cluster.  
4. Combine these components into a reduced dataset ready for modeling.

#### Why This Step Matters:
- Reduces **dimensionality** while maintaining data integrity.  
- Prevents **overfitting** caused by correlated predictors.  
- Speeds up training and improves model generalization.  
- Makes model interpretations more robust and clinically meaningful.

After this reduction, the dataset is optimized for downstream analysis — such as **classification (e.g., fetal health prediction)**, **clustering of patient profiles**, or **risk assessment modeling**.


```python
# Step 8A: Manual Feature Reduction (1 representative per cluster)

import pandas as pd

# Preview clusters from Step 7
print(cluster_map.sort_values("Cluster").head())

# Create a dictionary mapping each cluster → representative feature
# Here, we’ll simply pick the first feature from each cluster.
# You can manually edit this dict after inspecting feature importance later.
representatives = (
    cluster_map.groupby('Cluster')['Feature']
    .first()  # pick first; or use .apply(lambda x: x.iloc[0]) to customize
    .tolist()
)

print("\nSelected representative features per cluster:")
print(representatives)

# Create reduced dataset
df_reduced = df_clean[representatives].copy()

print(f"\nReduced dataset shape: {df_reduced.shape}")
display(df_reduced.head())
```

### Step 8B: PCA per Cluster (Automated Dimensionality Reduction)

This step automates feature reduction within each cluster using **Principal Component Analysis (PCA)**.  
The goal is to compress correlated variables into a smaller set of uncorrelated components while retaining most of the variance in the data.

#### Why PCA?
When multiple features within a cluster measure similar phenomena, they carry redundant information.  
PCA identifies the principal direction(s) of variance, summarizing the group’s information efficiently with fewer dimensions.

#### Process:
1. **Per-cluster standardization:**  
   Each cluster’s features are standardized (mean = 0, standard deviation = 1) to ensure equal scaling.
2. **Imputation of missing values:**  
   Any missing data are filled using the **median** of each feature to ensure PCA runs smoothly.
3. **PCA application:**  
   A single **Principal Component (PC)** is extracted from each cluster to represent it as one composite feature.  
   This component captures the maximum variance among the correlated variables.
4. **Variance explained:**  
   The percentage of total variance explained by the first PC is reported for each cluster.  
   - High explained variance (>60%) means strong correlation among variables (compact, clean signal).  
   - Lower explained variance (<40%) indicates diverse or loosely related variables, suggesting more PCs or manual review might be needed.




```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np

explained_target = 0.60   # change to 0.5 or 0.7 depending on how lossy you want
imputer = SimpleImputer(strategy='median')
results = pd.DataFrame(index=df_clean.index)

for cluster in sorted(cluster_map['Cluster'].unique()):
    features = cluster_map.loc[cluster_map['Cluster'] == cluster, 'Feature'].tolist()
    subset = df_clean[features]

    if subset.shape[1] == 0:
        print(f"Cluster {cluster}: no features, skipped")
        continue

    # Impute
    X = imputer.fit_transform(subset)

    # If only one feature -> standardized version
    if X.shape[1] == 1:
        col = StandardScaler().fit_transform(X).ravel()
        results[f'Cluster{cluster}_PC1'] = col
        print(f"Cluster {cluster}: single feature -> Cluster{cluster}_PC1")
        continue

    # Standardize
    Xs = StandardScaler().fit_transform(X)

    # Fit PCA with max components = number of features
    pca_full = PCA(n_components=min(Xs.shape[1], Xs.shape[0]-1))  # safe upper bound
    pca_full.fit(Xs)
    cumvar = np.cumsum(pca_full.explained_variance_ratio_)

    # choose number of components to reach target explained variance (or at least 1)
    n_comps = int(np.argmax(cumvar >= explained_target) + 1) if np.any(cumvar >= explained_target) else len(cumvar)
    n_comps = max(1, n_comps)

    # Transform with chosen n_comps
    pca = PCA(n_components=n_comps)
    comps = pca.fit_transform(Xs)  # shape (n_samples, n_comps)

    for i in range(n_comps):
        col_name = f'Cluster{cluster}_PC{i+1}'
        results[col_name] = comps[:, i]
    print(f"Cluster {cluster}: {X.shape[1]} features -> {n_comps} PCs (cum. explained {cumvar[n_comps-1]*100:.2f}%)")

df_reduced_auto = results.copy()
print("\nAuto PCA-reduced shape:", df_reduced_auto.shape)
display(df_reduced_auto.head())
```


#### Result:
- The final reduced dataset contains one column per cluster (e.g., `Cluster1_PC1`, `Cluster2_PC1`, etc.).  
- This dataset is much smaller (e.g., **4 columns instead of dozens**) but retains the key information content.
- It can now be used for downstream tasks such as **classification, regression, or further clustering.**

#### Notes:
- If a cluster shows **low explained variance**, consider:
  - Keeping more than one PC for that cluster (e.g., the first 2–3 PCs), or  
  - Selecting representative original features manually for interpretability.
- The resulting DataFrame (`df_reduced_pca`) is the input for **Step 9: Model Building & Evaluation.**

```python
# CLEAN NSP target, align X_df, and safe train_test_split
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings

# CONFIG
TARGET = 'NSP'
TOLERANCE = 0.15   # values within this distance to 1/2/3 will be snapped
TEST_SIZE = 0.2
RANDOM_STATE = 42

# Safety: require df_clean to exist
if 'df_clean' not in globals():
    raise RuntimeError("df_clean not found in globals. Load df_clean first (you already did earlier).")

# Prepare X_df if not present
if 'X_df' not in globals() or not isinstance(X_df, pd.DataFrame):
    X_df = df_clean.drop(columns=[TARGET], errors='ignore').copy()

# 1) Pull y_full, coerce numeric
y_full = pd.to_numeric(df_clean[TARGET], errors='coerce')
print("Original NSP value counts (top 30):")
print(df_clean[TARGET].value_counts(dropna=False).head(30))

# 2) Map noisy floats to nearest integer class {1,2,3} if within tolerance
def snap_to_classes(vals, classes=(1,2,3), tol=TOLERANCE):
    vals = pd.Series(vals)  # keep index
    snapped = vals.copy()
    snapped[:] = np.nan
    for k in classes:
        mask = vals.notna() & (vals.apply(lambda x: abs(float(x) - k) <= tol))
        snapped.loc[mask] = k
    return snapped

snapped = snap_to_classes(y_full, classes=(1,2,3), tol=TOLERANCE)

# 3) For exact integer matches (1,2,3) keep them
exact_mask = y_full.isin([1,2,3])
snapped.loc[exact_mask] = y_full.loc[exact_mask]

# 4) Build final cleaned y: if snapped has value use it; else if original was one of {1,2,3} use it; else NaN
y_clean = snapped.copy()
# any remaining exact integer values (in case floats equal exactly 1.0 etc)
y_clean[y_clean.isna() & exact_mask] = y_full[exact_mask]

# 5) Reporting: show what was snapped/removed
print("\nAfter snapping tolerance =", TOLERANCE)
print("Snapped value counts (non-null):")
print(y_clean.value_counts(dropna=True))

# Show which original values were mapped -> for debug
mapped_idx = y_clean[y_clean.notna()].index
unmapped_idx = y_clean[y_clean.isna()].index

print(f"\nRows total: {len(y_full)}. Mapped rows: {len(mapped_idx)}. Unmapped (to be dropped): {len(unmapped_idx)}")
if len(unmapped_idx) > 0:
    print("Examples of unmapped original NSP values (first 20):")
    print(y_full.loc[unmapped_idx].unique()[:20])

# 6) Drop unmapped (NaN) target rows
y = y_clean.dropna().astype(int).copy()
rows_kept = y.index
print(f"\nKeeping {len(rows_kept)} rows after dropping unmapped/invalid labels.")

# Subset X_df and df_clean to same rows
if isinstance(X_df, pd.DataFrame):
    X_df = X_df.loc[rows_kept].copy()
df_clean = df_clean.loc[rows_kept].copy()

print(f"Shapes after alignment -> X_df: {X_df.shape}, y: {y.shape}, df_clean: {df_clean.shape}")

# 7) Class sanity
class_counts = y.value_counts()
n_classes = class_counts.shape[0]
print("\nFinal class counts:")
print(class_counts.to_dict())

if n_classes < 2:
    raise RuntimeError(f"Only {n_classes} class present after cleaning target. Cannot proceed with supervised split.")

# 8) Decide if stratify is safe
def can_stratify(counts, test_size):
    # require at least 2 samples per class and at least one expected sample in test after floor
    if counts.min() < 2:
        return False
    expected_test = (counts * test_size).apply(np.floor)
    if (expected_test == 0).any():
        return False
    return True

use_stratify = can_stratify(class_counts, TEST_SIZE)
if not use_stratify:
    warnings.warn("Stratified split not possible (some classes too small). Falling back to random split.", UserWarning)

# 9) Perform split with graceful fallback
try:
    if use_stratify:
        X_train, X_test, y_train, y_test = train_test_split(
            X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
        )
    else:
        X_train, X_test, y_train, y_test = train_test_split(
            X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=None
        )
    print("\nTrain/test split successful.")
    print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
except Exception as e:
    # final fallback: try without stratify
    warnings.warn(f"train_test_split raised {e}. Retrying without stratify.", UserWarning)
    X_train, X_test, y_train, y_test = train_test_split(
        X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=None
    )
    print("\nTrain/test split successful (no stratify).")
    print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# 10) Expose objects to globals for notebook use
globals().update({
    'TARGET': TARGET,
    'X_df': X_df,
    'df_clean': df_clean,
    'y': y,
    'X_train': X_train,
    'X_test': X_test,
    'y_train': y_train,
    'y_test': y_test
})

print("\nDONE. TARGET set to", TARGET)
print("If you want a stricter or looser snapping tolerance, adjust TOLERANCE and re-run this cell.")
```

```python
# CLEAN + SNAP TARGET (NSP) -> SAFE STRATIFIED train_test_split
# Paste & run. Adjust TOLERANCE or explicit_remap as needed.
import numpy as np
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split

# ---------- CONFIG ----------
TARGET = 'NSP'
TOLERANCE = 0.15        # snap floats within this distance to nearest class (1,2,3)
TEST_SIZE = 0.2
RANDOM_STATE = 42
CLASSES = (1, 2, 3)
# Optional explicit remapping for known bad values (uncomment to use)
# explicit_remap = { -1.0: 3, 5.0: 2 }   # example: map -1 -> 3, 5 -> 2  (use with care)
explicit_remap = {}   # leave empty for now
SAVE_SPLITS = False   # set True to save X_train/X_test/y_train/y_test to CSVs
OUT_DIR = r"D:\DATA-SCIENCE\ASSIGNMENTS\5 EDA1"  # used if SAVE_SPLITS True
# ----------------------------

# Sanity checks
if 'df_clean' not in globals():
    raise RuntimeError("df_clean not found. Load the dataset into df_clean first.")

# Ensure X_df exists (features only)
if 'X_df' not in globals() or not isinstance(X_df, pd.DataFrame):
    X_df = df_clean.drop(columns=[TARGET], errors='ignore').copy()

# 1) Coerce numeric (safe)
y_full = pd.to_numeric(df_clean[TARGET], errors='coerce')

print("Original NSP value counts (top 30):")
print(df_clean[TARGET].value_counts(dropna=False).head(30))

# 2) Apply explicit remap first (if any)
if explicit_remap:
    # apply remap on a copy for inspection
    y_full = y_full.replace(explicit_remap)
    print("\nApplied explicit_remap:", explicit_remap)

# 3) Snap noisy floats to nearest integer classes (within tolerance)
def snap_to_classes(series, classes=CLASSES, tol=TOLERANCE):
    s = pd.Series(series)  # preserve index
    snapped = pd.Series(index=s.index, dtype=float)
    for k in classes:
        mask = s.notna() & (s.apply(lambda x: abs(float(x) - k) <= tol))
        snapped.loc[mask] = float(k)
    # keep exact integer-like values as well
    exact_mask = s.isin(classes)
    snapped.loc[exact_mask] = s.loc[exact_mask].astype(float)
    return snapped

snapped = snap_to_classes(y_full, classes=CLASSES, tol=TOLERANCE)

# 4) Final cleaned y: keep snapped values; others -> NaN (dropped later)
y_clean = snapped.copy()

# 5) Diagnostics before dropping
mapped_count = y_clean.notna().sum()
unmapped_count = y_clean.isna().sum()
print(f"\nAfter snapping tolerance={TOLERANCE}: mapped={mapped_count}, unmapped={unmapped_count}")

if unmapped_count > 0:
    print("Examples of unmapped original NSP values (first 20):")
    print(sorted(pd.Series(y_full.loc[y_clean[y_clean.isna()].index].unique())[:20]))

# 6) Drop unmapped (invalid) rows and align X_df & df_clean
keep_idx = y_clean.dropna().index
y = y_clean.loc[keep_idx].astype(int).copy()
X_df = X_df.loc[keep_idx].copy()
df_clean = df_clean.loc[keep_idx].copy()

print(f"\nShapes after alignment -> X_df: {X_df.shape}, y: {y.shape}, df_clean: {df_clean.shape}")

# 7) Class counts & check
class_counts = y.value_counts().sort_index()
print("\nFinal class counts:", class_counts.to_dict())
if class_counts.shape[0] < 2:
    raise RuntimeError(f"Only {class_counts.shape[0]} class present after cleaning. Cannot train a classifier.")

# 8) Stratify feasibility and split (fallback to non-stratified if needed)
def can_stratify(counts, test_size):
    if counts.min() < 2:
        return False
    expected_test = (counts * test_size).apply(np.floor)
    return (expected_test > 0).all()

use_stratify = can_stratify(class_counts, TEST_SIZE)
if not use_stratify:
    warnings.warn("Stratified split not possible (some classes too small). Falling back to random split.", UserWarning)

try:
    if use_stratify:
        X_train, X_test, y_train, y_test = train_test_split(
            X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
        )
    else:
        X_train, X_test, y_train, y_test = train_test_split(
            X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=None
        )
    print("\nTrain/test split successful.")
    print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
except Exception as exc:
    warnings.warn(f"train_test_split failed: {exc}. Retrying without stratify.", UserWarning)
    X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=None)
    print("\nTrain/test split successful (fallback).")
    print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# 9) Optional: save splits
if SAVE_SPLITS:
    import os
    os.makedirs(OUT_DIR, exist_ok=True)
    X_train.to_csv(os.path.join(OUT_DIR, "X_train.csv"), index=True)
    X_test.to_csv(os.path.join(OUT_DIR, "X_test.csv"), index=True)
    y_train.to_csv(os.path.join(OUT_DIR, "y_train.csv"), index=True)
    y_test.to_csv(os.path.join(OUT_DIR, "y_test.csv"), index=True)
    print(f"\nSaved splits to {OUT_DIR}")

# 10) Expose to globals
globals().update({
    'TARGET': TARGET,
    'X_df': X_df,
    'df_clean': df_clean,
    'y': y,
    'X_train': X_train,
    'X_test': X_test,
    'y_train': y_train,
    'y_test': y_test
})

print("\nDONE — TARGET set to", TARGET)
print("Tip: if you want -1 or 5 mapped explicitly (instead of dropped), set explicit_remap = { -1.0: 3, 5.0: 2 } and re-run.")
```
